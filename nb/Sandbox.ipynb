```python
from jupyter_dash import JupyterDash
from dash import Dash, dcc, html, Input, Output
import plotly.graph_objects as go
import numpy as np

# Initialize the app
app = JupyterDash(__name__)
```

```python
from rail.estimation.algos import bpz_lite
from rail.utils import catalog_utils

model_file = 'pz_challenge_taskset_1_cardinal_pz_model_10yr.pkl'
```

```python
catalog_utils.load_yaml("catalogs.yaml")
catalog_utils.apply("cardinal_roman_rubin")

estimator = bpz_lite.BPZliteEstimator.make_stage(
    name=f"estimate",
    model=model_file,
    output_mode="return",
    spectra_file="COSMOS_seds.list",
)
estimator.open_model(**estimator.config)
estimator.validate()
estimator._initialize_run()
```

```python
# app.py (or run in notebook with jupyter-dash)

# Initialize the Dash app
app = Dash(__name__)

# Define the layout
app.layout = html.Div([
    html.H1("PDF Explorer"),
    
    html.Div([
        html.Div([
            html.Label('mag_u:'),
            dcc.Slider(id='mag_u', min=20, max=25, step=0.1, value=22.0, 
                      marks={i: str(i) for i in range(20, 26)},
                      tooltip={"placement": "bottom", "always_visible": True},
                      updatemode='drag') # Add this line         
        ], style={'margin': '10px'}),
        
        html.Div([
            html.Label('mag_g:'),
            dcc.Slider(id='mag_g', min=20, max=25, step=0.1, value=22.0,
                      marks={i: str(i) for i in range(20, 26)},
                      tooltip={"placement": "bottom", "always_visible": True},
                      updatemode='drag')  # Add this line                     
        ], style={'margin': '10px'}),
        
        html.Div([
            html.Label('mag_r:'),
            dcc.Slider(id='mag_r', min=20, max=25, step=0.1, value=22.0,
                      marks={i: str(i) for i in range(20, 26)},
                      tooltip={"placement": "bottom", "always_visible": True},
                      updatemode='drag')  # Add this line         
        ], style={'margin': '10px'}),
        
        html.Div([
            html.Label('mag_i:'),
            dcc.Slider(id='mag_i', min=20, max=25, step=0.1, value=22.0,
                      marks={i: str(i) for i in range(20, 26)},
                      tooltip={"placement": "bottom", "always_visible": True},
                      updatemode='drag')  # Add this line                     
        ], style={'margin': '10px'}),
        
        html.Div([
            html.Label('mag_z:'),
            dcc.Slider(id='mag_z', min=20, max=25, step=0.1, value=22.0,
                      marks={i: str(i) for i in range(20, 26)},
                      tooltip={"placement": "bottom", "always_visible": True},
                      updatemode='drag')  # Add this line                     
        ], style={'margin': '10px'}),
        
        html.Div([
            html.Label('mag_y:'),
            dcc.Slider(id='mag_y', min=20, max=25, step=0.1, value=22.0,
                      marks={i: str(i) for i in range(20, 26)},
                      tooltip={"placement": "bottom", "always_visible": True},
                      updatemode='drag')  # Add this line                     
        ], style={'margin': '10px'}),
        
        html.Div([
            html.Label('mag_Y:'),
            dcc.Slider(id='mag_Y', min=20, max=25, step=0.1, value=22.0,
                      marks={i: str(i) for i in range(20, 26)},
                      tooltip={"placement": "bottom", "always_visible": True},
                      updatemode='drag')  # Add this line                     
        ], style={'margin': '10px'}),
        
        html.Div([
            html.Label('mag_J:'),
            dcc.Slider(id='mag_J', min=20, max=25, step=0.1, value=22.0,
                      marks={i: str(i) for i in range(20, 26)},
                      tooltip={"placement": "bottom", "always_visible": True},
                      updatemode='drag')  # Add this line                     
        ], style={'margin': '10px'}),
        
        html.Div([
            html.Label('mag_H:'),
            dcc.Slider(id='mag_H', min=20, max=25, step=0.1, value=22.0,
                      marks={i: str(i) for i in range(20, 26)},
                      tooltip={"placement": "bottom", "always_visible": True},
                      updatemode='drag')  # Add this line                     
        ], style={'margin': '10px'}),
    ], style={'width': '48%', 'display': 'inline-block', 'vertical-align': 'top'}),
    
    html.Div([
        dcc.Graph(id='pdf-plot')
    ], style={'width': '48%', 'display': 'inline-block'}),
])

first = True

# Callback to update the plot
@app.callback(
    Output('pdf-plot', 'figure'),
    [Input('mag_u', 'drag_value'),
     Input('mag_g', 'drag_value'),
     Input('mag_r', 'drag_value'),
     Input('mag_i', 'drag_value'),
     Input('mag_z', 'drag_value'),
     Input('mag_y', 'drag_value'),
     Input('mag_Y', 'drag_value'),
     Input('mag_J', 'drag_value'),
     Input('mag_H', 'drag_value')]
)
def update_plot(mag_u, mag_g, mag_r, mag_i, mag_z, mag_y, mag_Y, mag_J, mag_H):
    dd = dict(
        mag_u_lsst=np.array([mag_u]),
        mag_g_lsst=np.array([mag_g]),
        mag_r_lsst=np.array([mag_r]),
        mag_i_lsst=np.array([mag_i]),
        mag_z_lsst=np.array([mag_z]),
        mag_y_lsst=np.array([mag_y]),
        mag_Y_roman=np.array([mag_Y]),
        mag_J_roman=np.array([mag_J]),
        mag_H_roman=np.array([mag_H]),    
        mag_u_lsst_err=np.array([0.1]),
        mag_g_lsst_err=np.array([0.1]),
        mag_r_lsst_err=np.array([0.1]),
        mag_i_lsst_err=np.array([0.1]),
        mag_z_lsst_err=np.array([0.1]),
        mag_y_lsst_err=np.array([0.1]),
        mag_Y_roman_err=np.array([0.1]),
        mag_J_roman_err=np.array([0.1]),
        mag_H_roman_err=np.array([0.1]),        
    )
    
    estimator.data_store.clear()
    estimator._process_chunk(0, 1, dd, first)
    handle = estimator.get_handle('output')
    
    xvals = np.linspace(0., 3., 301)
    yvals = np.squeeze(handle.data.pdf(xvals))
    
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=xvals, y=yvals, mode='lines'))
    fig.update_layout(
        title='PDF',
        xaxis_title='X',
        yaxis_title='Y',
        hovermode='x unified'
    )
    
    return fig
```
