```python
import numpy as np
import plotly.graph_objects as go
from ipywidgets import interactive, FloatSlider, VBox
from IPython.display import display
```

```python
def plot_sine(frequency=1.0, amplitude=1.0, phase=0.0):
    """
    Plot a sine wave with adjustable frequency, amplitude, and phase.
    """
    # Create x values
    x = np.linspace(0, 4*np.pi, 1000)
    
    # Calculate y values
    y = amplitude * np.sin(2 * np.pi * frequency * x / (2*np.pi) + phase)
    
    # Create the plot
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', line=dict(color='blue', width=2)))
    
    fig.update_layout(
        title=f'Sine Wave: y = {amplitude:.2f} × sin({frequency:.2f}x + {phase:.2f})',
        xaxis_title='x (radians)',
        yaxis_title='y',
        yaxis_range=[-5, 5],
        height=500,
        showlegend=False,
        hovermode='x'
    )
    
    fig.show()
```

```python
# Create interactive sliders
frequency=FloatSlider(min=0.1, max=5.0, step=0.1, value=1.0, description='Frequency:', continuous_update=True)
amplitude=FloatSlider(min=0.1, max=5.0, step=0.1, value=1.0, description='Amplitude:', continuous_update=True)
phase=FloatSlider(min=0, max=2*np.pi, step=0.1, value=0.0, description='Phase (rad):', continuous_update=True)

w = interactive(
    plot_sine,
    frequency=frequency,
    amplitude=amplitude,
    phase=phase,
)
```

```python
# Display the widget
display(w)
```
