```python
import os
import tables_io
import qp

from rail.utils import catalog_utils
from rail.estimation.algos import bpz_lite

from live_rail.wrappers.cat_estimator_pdf_wrapper import CatEstimatorPdfWrapper
from live_rail.apps.cat_data_app import build_data_app
```

```python
data_file = 'public/pz_challenge_taskset_1_cardinal_training_10yr.hdf5'
```

```python
data = tables_io.read(data_file)
```

```python
model_file = 'pz_challenge_taskset_1_cardinal_pz_model_10yr.pkl'

catalog_utils.load_yaml("catalogs.yaml")
catalog_utils.apply("cardinal_roman_rubin")

wrapper = CatEstimatorPdfWrapper.build_wrapper(
    bpz_lite.BPZliteEstimator,
    n_obj=100000,
    name="estimate",
    model=model_file,
    output_mode="return",
    spectra_file="COSMOS_seds.list",
)    
```

```python
if not os.path.exists('pz_estimate.hdf5'):
    pz_data = wrapper(data)
    pz_data.write_to('pz_estimate.hdf5')
else:
    pz_data = qp.read('pz_estimate.hdf5')
```

```python
app = build_data_app(data, pz_data)
```

```python
app.run(mode='inline', port=8051)
```
